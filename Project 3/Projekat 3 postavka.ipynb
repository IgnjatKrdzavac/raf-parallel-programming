Koristeći PyCuda okurženje napisati CUDA program za (matrično) množenje matrica. Potrebno je CUDA kernel zadovolji Blas interfejs za generalizovano mnozenje matrica: 
```
gemm(transa, transb, m, n, k, alpha, A, B): 
     alpha * (opA(A) X opB(B))
```
pri čemu su:
- `transa` i `transb`: char parametri koji definišu operacije opA i opB, na sledeći način:
  - `'N': op(X) = X` - (ne menja matricu)
  - `'T': op(X) = X<sup>T</sup>` (transponuje matricu)
- `m` - broj redova matrice A
- `n` - broj kolona matrice A
- `k` - dimenzija matrice B koja nedostaje (u zavisnosti od transa i transb može biti broj redova ili broj kolona) 
- `alpha` - skalarna vrednost (float) kojim se množe svi elementi rezultujuće matrice
- `A` i `B` - pokazivači na matrice A i B  

Transponovanje je potrebno ugraditi u postupak mnozenja (u slučaju da je potrebno transponovati neku matricu, to ne treba implementirati kao poseban korak transponovanja pre mnozenja, već izmenom indeksa u izvornim matricama kojima sa pritupa kod racunanja rezultata).  

Provere da li su matrice odgovarajućih dimenzija kako bi se operacija mogla izvršiti treba izvršiti u python kodu, pre alokacije memorije na CUDA uređaju. 


1. [5 bodova] Program koji vrši množenja matrica malih dimenzija (množenje se može izvršiti upotrebom jednom bloka niti)
2. [7 bodova] Probram koji vrši množenje matrica većih dimenzija, upotrebom većeg broja CUDA blokova.
3. [8 bodova] Ubrzati rešenje iz stavke 2 upotrebom deljene memorije (tako da niti jednog bloka prvo dovuku deo podataka u deljenu memeorju, a potom sve čitaju iz deljene memorije)
